**29.  Write a code to combine two dictionaries into one and print the result?**

```python

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

combined_dict_unpacking = {**dict1, **dict2}
print(f"Combined dictionary using unpacking: {combined_dict_unpacking}")
combined_dict_update = dict1.copy()
combined_dict_update.update(dict2)
print(f"Combined dictionary using update(): {combined_dict_update}")
```
